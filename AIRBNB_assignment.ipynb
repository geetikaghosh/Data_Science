```python
#check for duplicate data
df= df.drop_duplicates()
df.count()
```

```python
#check null values in columns

df.isnull().sum()
```

```python
#fill name and host_name with a substitute value

df['name'].fillna('unknown', inplace= True)
df['host_name'].fillna('NA', inplace=True)
df.isnull().sum()
```

```python
#replace 0 to the null values in the columns that has most null 

df['reviews_per_month']= df['reviews_per_month'].replace(to_replace= np.nan, value=0).astype('int64')
df.info()
```

```python
df = df.drop(['last_review'], axis=1)
df.info()
```

```python
sns.boxplot(x = df['price'])
plt.show()
```

```python
#(1) Distribution Of Airbnb Bookings Price Range

plt.figure(figsize=(8,6))
sns.histplot(df[df['price'] <= 1000], x='price', bins="doane", kde=False, color='cyan')
plt.title('Distribution Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#(2) Total Listing/Property count in Each Neighborhood Group

counts = df['neighbourhood_group'].value_counts()
Top_Neighborhood_group = counts.reset_index()
Top_Neighborhood_group.columns = ['Neighborhood_Groups', 'Listing_Counts']
Top_Neighborhood_group
```

```python

plt.figure(figsize=(7,5))
sns.countplot(x=df['neighbourhood_group'])
```

```python
#(3) Average Price Of Each Neighborhood Groups

#import mean function from the statistics module 
from statistics import mean
sns.pointplot(x = 'neighbourhood_group', y='price', data=df, estimator = np.mean)
```

```python
#(4) Top Neighborhoods by Listing/property

top_10_neigbourhoods = df['neighbourhood'].value_counts().nlargest(10)
colors = ['c', 'green', 'olive', 'y', 'm', 'orange', 'pink', 'cyan', 'red', 'blue']
top_10_neigbourhoods.plot(kind='bar', figsize=(15, 6), color = colors)
```

```python
#(5) Number Of Active Hosts Per Location

hosts_per_location = df.groupby('neighbourhood_group')['id'].count()
locations = hosts_per_location.index
host_counts = hosts_per_location.values
plt.figure(figsize=(12, 5))
plt.plot(locations, host_counts, marker='o', ms=12, mew=4, mec='r')
plt.show()
```

```python
#(6) Total Counts Of Each Room Type

room_type_counts = df['room_type'].value_counts()

labels = room_type_counts.index
sizes = room_type_counts.values

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#(7) Find stay Requirement counts by Minimum Nights

min_nights_count = df.groupby('minimum_nights').size().reset_index(name = 'count')
minimum_nights = min_nights_count['minimum_nights']
count = min_nights_count['count']
plt.figure(figsize=(30, 5))
plt.bar(minimum_nights, count)
plt.show()
```

```python
#(8) Maximum Reviews by Each Neighborhood Group

plt.figure(figsize=(10, 6))
sns.histplot(df[df['number_of_reviews'] <= 100], x='number_of_reviews', binwidth=5, kde=False, color='blue')
plt.title('Distribution of Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.xticks(np.arange(0, 101, 5))
plt.yticks(np.arange(0, 24001, 2000))
plt.show()
```

```python
#(9) most reviewed room type per month in neighbourhood groups

f, ax = plt.subplots(figsize=(8, 8))

ax = sns.stripplot(x='room_type', y='reviews_per_month', hue='neighbourhood_group', dodge=True, data=df, palette='Set1')

ax.set_title('Most Reviewed room_types in each Neighbourhood Groups', fontsize='14')
```

```python
#(10) Find Best Location Listing/Property Location For Host

plt.figure(figsize=(6, 4))
ax = sns.histplot(df[df['calculated_host_listings_count'] <= 10], x='calculated_host_listings_count', binwidth=1, kde=False, color='cyan')
plt.title('Distribution of Properties per Host')
plt.xlabel('Number of Properties')
plt.ylabel('Frequency')
plt.xticks(np.arange(0, 11, 1))
plt.show()
```

```python
#(11) Find Best Location Listing/Property Location For Travellers

plt.figure(figsize=(12, 4))
ax = sns.histplot(df[df['calculated_host_listings_count'] <= 10], x='calculated_host_listings_count', binwidth=1, kde=False, color='m')
plt.title('Distribution of Properties per Host')
plt.xlabel('Number of Properties')
plt.ylabel('Frequency')
plt.xticks(np.arange(0, 11, 1))
plt.show()
```

```python
#(12) Find Price Variation in NYC Neighborhood groups

plt.figure(figsize=(10, 6))
sns.histplot(df[df['price'] <= 1000], x='price', bins="doane", kde=False, color='blue')
plt.title('Distribution of Rental Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
